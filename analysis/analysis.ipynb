```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams.update({'font.size': 22})
```

```python
adj_list = pd.read_csv("adjacency_list_eulerian.csv")
adj_list["name"] = "Lista de adjacência"
adj_matrix = pd.read_csv("adjacency_matrix_eulerian.csv")
adj_matrix["name"] = "Matriz de Adjacência"
inc_matrix = pd.read_csv("incidency_matrix_eulerian.csv")
inc_matrix["name"] = "Matriz de Incidência"
```

```python
df = pd.concat([adj_list, adj_matrix, inc_matrix]).reset_index(drop=True)
```

```python
df
```

```python
fig = plt.figure(figsize=(20,10))
sns.lineplot(data=df, x="v", y="op", hue="name", linewidth=3)
plt.yscale("log")
plt.title("Circuito Euleriano", fontsize=30, fontweight="bold")
plt.xlabel("Número de vértices", fontsize=25)
plt.ylabel("Número de operações", fontsize=25)
plt.legend(title='Estrutura de dados', loc='upper left', fontsize=20)
```
